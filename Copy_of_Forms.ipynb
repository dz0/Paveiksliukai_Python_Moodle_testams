```
# http://geopandas.org/mapping.html
# !pip install geopandas # if needed to install
import geopandas as gpd

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
# from bokeh.sampledata.airport_routes import airports


world = world[(world.pop_est>0) & (world.name!="Antarctica")] # we won't fly to some places for now
world['gdp_per_cap'] = world.gdp_md_est / world.pop_est

plt_base = world.plot(column='gdp_per_cap', cmap='OrRd');
cities.plot(ax=plt_base, marker='o', color='green', markersize=5);
for x, y, label in zip(cities.geometry.x, cities.geometry.y, cities.name):
    plt_base.annotate(label[:3], xy=(x, y), xytext=(3, 3), textcoords="offset points", c="b", alpha=0.5)
# for arrows https://matplotlib.org/tutorials/text/annotations.html#using-connectionpatch
# https://matplotlib.org/tutorials/text/annotations.html#annotating-with-arrow
```
